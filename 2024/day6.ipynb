```python
inputt = '''....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...'''
```

```python
inputt = inputt.split('\n')
inputt = [list(i) for i in inputt]
```

```python
r = len(inputt)
c = len(inputt[0])
```

```python
cr,cc=0,0
```

```python
for i in range(r):
    for j in range(c):
        if inputt[i][j] in ['^','v','<','>']:
            cr,cc = i,j
            break
```

```python
while cr<r-1 and cr>0 and cc<c-1 and cc>0:
    if inputt[cr][cc] == '^':
        inputt[cr][cc] = 'X'
        if inputt[cr-1][cc] == '#':
            inputt[cr][cc] = '>'
        else:
            cr-=1
            inputt[cr][cc] = '^'
        # print(cr,cc)
    elif inputt[cr][cc] == 'v':
        inputt[cr][cc] = 'X'
        if inputt[cr+1][cc] == '#':
            inputt[cr][cc] = '<'
        else:
            cr+=1
            inputt[cr][cc] = 'v'
    elif inputt[cr][cc] == '<':
        inputt[cr][cc] = 'X'
        if inputt[cr][cc-1] == '#':
            inputt[cr][cc] = '^'
        else:
            cc-=1
            inputt[cr][cc] = '<'
    elif inputt[cr][cc] == '>':
        inputt[cr][cc] = 'X'
        if inputt[cr][cc+1] == '#':
            inputt[cr][cc] = 'v'
        else:
            cc+=1
            inputt[cr][cc] = '>'
    # print(cr,cc)
```
